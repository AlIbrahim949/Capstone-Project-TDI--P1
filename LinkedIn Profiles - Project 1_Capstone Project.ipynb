```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
import datetime
sns.set()
```

```python
data=pd.read_csv('F:/The Data Incubator/temp_datalab_records_linkedin_company.csv')
```

```python
new_data = data.copy()
new_data
```

```python
Oracle_data=new_data.loc[new_data['company_name'] == 'Oracle'].copy()
Oracle_data
```

```python
from datetime import datetime 
```

```python
Oracle_data['date'] = pd.DatetimeIndex(Oracle_data['date_added']).date
Oracle_data
```

```python
x= Oracle_data['date']
y=Oracle_data['employees_on_platform']
plt.scatter(x,y)
plt.xlabel('Date', fontsize=20)
plt.ylabel('Number of Employees on Platform', fontsize=20)
```

```python
Exxon_data=new_data.loc[new_data['company_name'] == 'ExxonMobil'].copy()
Exxon_data
```

```python
Exxon_data['date'] = pd.DatetimeIndex(Exxon_data['date_added']).date
Exxon_data
```

```python
x= Exxon_data['date']
y=Exxon_data['employees_on_platform']
plt.scatter(x,y)
plt.xlabel('Date', fontsize=20)
plt.ylabel('Number of Employees on Platform', fontsize=20)
```

```python
x= Exxon_data['date']
y=Exxon_data['followers_count']
plt.scatter(x,y)
plt.xlabel('Date', fontsize=20)
plt.ylabel('Number of Employees on Platform', fontsize=20)
```
