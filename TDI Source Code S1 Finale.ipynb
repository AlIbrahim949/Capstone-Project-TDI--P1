# Question 5 - How many arrest incidents occurred within 2 km from the Bradbury Building in 2018?

```python
q5_eighteen_data = eighteen_data.copy()
q5_eighteen_nonull_data = q5_eighteen_data[q5_eighteen_data['Location'] != (0,0)]
```

```python
split_data = q5_eighteen_nonull_data.Location.str.strip(')').str.strip('(').str.split(', ')
q5_eighteen_nonull_data['lat'] = split_data.apply(lambda x: x[0])
q5_eighteen_nonull_data['long'] = split_data.apply(lambda x: x[1])
```

```python
lat_q5=q5_eighteen_nonull_data['lat'].to_numpy()
long_q5=q5_eighteen_nonull_data['long'].to_numpy()
lat_f = lat_q5.astype(np.float)
long_f = long_q5.astype(np.float)
```

```python
import math
pi=math.pi
pi
```

```python
lat_f = (lat_f* pi)/180
long_f = (long_f*pi)/180
delt_lat = lat_f - 0.594293965
mean_lat = (lat_f +0.594293965)/2 
delt_long = long_f - (-2.063814508)
cos_mean_lat = np.cos(mean_lat)
delt_lat , delt_long, mean_lat,cos_mean_lat
```

```python
R = 6371
f = ((delt_lat**2)+((cos_mean_lat*delt_long)**2))
D = R* (f**0.5)
D
```

```python
q5_eighteen_nonull_data['Distance'] = D
q5f_eighteen_nonull_data=q5_eighteen_nonull_data.loc[q5_eighteen_nonull_data['Distance'] <= 2].copy()
```

# Question 7 - Report the average of the top 5 of the calculated ratio

```python
q7_11to18_data=new_date_data.loc[new_date_data['year'] != 2019].copy()
q7_11to18_data.dropna(subset=['Charge Group Code'],inplace=True)
q7_11to18_data
```

```python
q7_11to18_final_data = q7_11to18_data.groupby(['Charge Group Code','Area ID']).size().reset_index(name='counter')
q7_11to18_final_data
```

```python
q7_11to18_final_data=q7_11to18_final_data.loc[q7_11to18_final_data['Charge Group Code'] != 99].copy()
q7_11to18_final_data
```

```python
q7_total_GC = q7_11to18_final_data.groupby('Charge Group Code')['counter'].sum().reset_index(name='GC counter')
q7_total_GC
```

```python
q7_total = q7_total_GC['GC counter'].sum()
q7_total
```

```python
q7_total_GC['GC Prob' ] = q7_total_GC['GC counter']/q7_total
q7_total_GC
```

```python
q7_11to18_final_data = q7_11to18_final_data.merge( q7_total_GC, on=['Charge Group Code'], how='inner')
q7_11to18_final_data
```

```python
q7_11to18_final_data ['Area Prob'] = q7_11to18_final_data['counter']/q7_11to18_final_data['GC counter']

q7_11to18_final_data
```

```python
q7_11to18_final_data ['Total Ratio'] = q7_11to18_final_data['Area Prob']/q7_11to18_final_data['GC Prob']
q7_11to18_final_data
```

```python
from heapq import nlargest
q = q7_11to18_final_data ['Total Ratio'].copy()
large_5 = nlargest(5,q)
large_5
```

```python
import statistics
average_ratio = statistics.mean(large_5)
average_ratio
```

# Question 8 - What is the projected number of felony arrests in 2019? 

# Pre-Deliquency, Non- Criminal Detention and Nan due to uncertainty all treated as misdemeanors and hence neglected.

```python
mask = q7_11to18_data['Charge Group Description'].isin(['Pre-Delinquency','Non-Criminal Detention'])
q7_11to18_data.dropna(subset=['Charge Group Description'])
q7_11to18_data.dropna(subset=['Charge Group Description'],inplace=True)
q8_felony_data = q7_11to18_data[~mask].copy()
```

```python
q8_felony_tally = q8_felony_data.groupby(['year','Charge Group Code']).size().reset_index(name='counter')
q8_final_tally = q8_felony_tally.groupby('year')['counter'].sum().reset_index(name='Number of Felonies')
```

```python
x= q8_final_tally['year']
y=q8_final_tally['Number of Felonies']

plt.scatter(x,y)
plt.xlabel('Year', fontsize=20)
plt.ylabel('Number of Felonies', fontsize=20)
```

```python
x_matrix = x.values.reshape(-1,1)
x_matrix.shape
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_matrix,y)
```
